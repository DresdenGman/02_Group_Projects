### viewdata

```python
import pandas as pd
from datetime import datetime

data=pd.read_csv("order-14.1.csv",parse_dates=["time"],encoding='gbk')
data.head()
# print(data.head(5))
# viewdata
data.info()
```

### calculatemonthrelated 

```python
# calculatemonthrelated 
This_month=data[(data["time"]>=datetime(2018,2,1))&(data["time"]<=datetime(2018,2,28))]
# Salescalculate
sales_1=(This_month[""]*This_month['']).sum()
# calculate
traffic_1=This_month["OrderID"].drop_duplicates().count()
# calculate
s_t_1=sales_1/traffic_1
print("monthSales：{:.2f},：{},：{:.2f}".format(sales_1,traffic_1,s_t_1))
```

### calculatemonthrelated 

```python
# calculatemonthrelated 
last_month=data[(data["time"]>=datetime(2018,1,1))&(data["time"]<=datetime(2018,1,31))]

# Salescalculate
sales_2=(last_month[""]*last_month['']).sum()
# calculate
traffic_2=last_month["OrderID"].drop_duplicates().count()
# calculate
s_t_2=sales_2/traffic_2
print("monthSales：{:.2f},：{},：{:.2f}".format(sales_2,traffic_2,s_t_2))
```

### calculateyearrelated 

```python
# calculateyearrelated 
same_month=data[(data["time"]>=datetime(2017,2,1))&(data["time"]<=datetime(2017,2,28))]

sales_3=(same_month[""]*same_month[""]).sum()

traffic_3=same_month["OrderID"].drop_duplicates().count()
s_t_3=sales_3/traffic_3
print("monthSales：{:.2f},：{},：{:.2f}".format(sales_3,traffic_3,s_t_3))
```

```python
# utilizing encoding
def get_month_data(data):
    sale=(data[""]*data[""]).sum()
    traffic=data["OrderID"].drop_duplicates().count()
    s_t=sale/traffic
    return (sale,traffic,s_t)

# monthrelated 
sales_1,traffic_1,s_t_1=get_month_data(This_month)
print(sales_1,traffic_1,s_t_1)

# monthrelated 
sales_2,traffic_2,s_t_2=get_month_data(last_month)
print(sales_2,traffic_2,s_t_2)

# yearrelated 
sales_3,traffic_3,s_t_3=get_month_data(same_month)
print(sales_3,traffic_3,s_t_3)
```

### DataFrame  and 

```python
# DataFrame
report=pd.DataFrame([[sales_1,sales_2,sales_3],[traffic_1,traffic_2,traffic_3],[s_t_1,s_t_2,s_t_3]],
                    columns=["month","month","year"],index=["Sales","",""])
# print(report)
#  and 
report[""]=report["month"]/report["month"]-1

report[""]=report["month"]/report["year"]-1
```

```python
# 
color_js = """new echarts.graphic.LinearGradient(0, 0, 1, 0,
    [{offset: 0, color: '#009ad6'}, {offset: 1, color: '#ed1941'}], false)"""


sort_info = df.sort_values(by='', ascending=True)
b1 = (
    Bar()
    .add_xaxis(list(sort_info[''])[-20:])
    .add_yaxis('Attraction', sort_info[''].values.tolist()[-20:],itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_js)))
    .reversal_axis()
    .set_global_opts(
        title_opts=opts.TitleOpts(title='Attractiondata'),
        yaxis_opts=opts.AxisOpts(name='Attraction'),
        xaxis_opts=opts.AxisOpts(name=''),
        )
    .set_series_opts(label_opts=opts.LabelOpts(position="right"))

)
# 
g1 = (
    Grid()
        .add(b1, grid_opts=opts.GridOpts(pos_left='20%', pos_right='5%'))  
)
g1.render_notebook()
```

## 3.3 Provincial4A-5ACount

```python
# 
color_js = """new echarts.graphic.LinearGradient(0, 1, 0, 0,
    [{offset: 0, color: '#009ad6'}, {offset: 1, color: '#ed1941'}], false)""" 

df_tmp2 = df[df[''].isin(['4A', '5A'])]
df_counts = df_tmp2.groupby('City').count()['']
b2 = (
        Bar()
            .add_xaxis(df_counts.index.values.tolist())
            .add_yaxis('4A-5ACount', df_counts.values.tolist(),itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_js)))
            .set_global_opts(
            title_opts=opts.TitleOpts(title='Provincial4A-5ACount'),
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_='inside')],
        )
    )
b2.render_notebook()
```

## 3.7 PriceShare

```python
# PriceShare
price_level = [0, 50, 100, 150, 200, 250, 300, 350, 400, 500]    
label_level = ['0-50', '50-100', '100-150', '150-200', '200-250', '250-300', '300-350', '350-400', '400-500']    
jzmj_cut = pd.cut(df['Price'], price_level, labels=label_level)        
df_price = jzmj_cut.value_counts()
df_price
```

```python
p1 = (
    Pie(init_opts=opts.InitOpts(
            width='800px', height='600px',
            )
       )
        .add(
        '',
        [list(z) for z in zip(df_price.index.tolist(), df_price.values.tolist())],
        radius=['20%', '60%'],
        center=['40%', '50%'],
        rosetype='radius',
        label_opts=opts.LabelOpts(is_show=True),
        )    
        .set_global_opts(title_opts=opts.TitleOpts(title='PriceShare',pos_left='33%',pos_top="5%"),
                        legend_opts=opts.LegendOpts(type_='scroll', pos_left="80%",pos_top="25%",orient="vertical")
                        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}: {c} ({d}%)'),position='outside')
    )
p1.render_notebook()
```

## 3.9 Attraction

```python
contents = "".join('%s' % i for i in df[''].values.tolist())
contents_list = jieba.cut(contents)
ac = Counter(contents_list)

stopwords = []
with open('./stopwords.txt', "r",encoding='utf-8') as f:  # file
    data = f.read()  # file
    stopwords = data.split('\n')

for i in stopwords:
    del ac[i]
 
w1 = (
    WordCloud()
    .add("", 
         ac.most_common(150), 
         word_size_range=[5, 100], 
         textstyle_opts=opts.TextStyleOpts(font_family="cursive"),
         shape='star')
    .set_global_opts(title_opts=opts.TitleOpts(title="Attraction"))
)
w1.render_notebook()
```
